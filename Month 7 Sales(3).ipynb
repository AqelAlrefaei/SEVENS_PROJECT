## 1- زيادة متوسط الربح لكل سيارة

```python
avgProfit = df.groupby("النوع")["الربح"].mean().reset_index()
avgProfit = avgProfit.rename(columns={"الربح": "متوسط الربح"})
```

### متوسط الربح قبل الزياده

```python
avgProfit["متوسط الربح"] 
```

### متوسط الربح بعد زياده 1000 ريال

```python
avgProfit["متوسط الربح بعد الزيادة"] = avgProfit["متوسط الربح"] + 1000
```

```python
avgProfit['متوسط الربح بعد الزيادة']
```

```python
df = df.merge(avgProfit, on="النوع", how="left")
```

```python
df[["ماركة السيارة","النوع", "السعر ش.الضريبة", "الربح", "متوسط الربح بعد الزيادة"]]
```

### 3,4 - تحسين نسبة تسليم السيارات المباعة و تتبع نسبة الصفقات التي لم يتم تسليمها

```python
sold = (df['تم البيع'] == True).sum()
total = len(df)
sold_percent = (sold / total) * 100
print(f"نسبة السيارات المباعة: {sold_percent:.2f}%")
```

```python
delivered = (df['تم التسليم'] == True).sum()
undelivered = (df['تم التسليم'] != True).sum()
print("عدد السيارات المسلمه:", delivered)
print("عدد السيارات الغير مسلمه:",undelivered)
```

```python
import matplotlib.pyplot as plt

sold = (df['تم التسليم'] == True).sum()
not_sold = (df['تم التسليم'] == False).sum()

values = [sold, not_sold]
labels = ['delivered', 'undelivered']

plt.figure(figsize=(6,6))
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=90, colors=['#4CAF50','#F44336'])
plt.title("Percentage of delivered cars vs. undelivered cars")
plt.show()
```

```python
Image("Percentage of delivered cars vs. undelivered cars.png") # تم وضعها كصوره بسبب مشكله في التوافق على جهازي
```

## لتحسين نسبة التسليم قمنا بتسليم 20 سيارة بحسب تاريخ الطلب (من أقدم طلب لم يتم تسليمه)

```python
false_index = df[df["تم التسليم"] == False].index
df.loc[false_index[:20], "تم التسليم"] = True
```

```python
import matplotlib.pyplot as plt

sold = (df['تم التسليم'] == True).sum()
not_sold = (df['تم التسليم'] == False).sum()

values = [sold, not_sold]
labels = ['delivered', 'undelivered']

plt.figure(figsize=(6,6))
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=90, colors=['#4CAF50','#F44336'])
plt.title("Percentage of delivered cars vs. undelivered cars")
plt.show()
```

```python
Image('Percentage of delivered cars vs. undelivered cars after improving the ratio by delivering 20 cars.png')
```

### مقارنه اداء كل جهه تمويل من حيث عدد المبيعات والربحية

```python
sales_count2 = df[df["تم التسليم"] == True].groupby("جهة التمويل")["تم البيع"].count()

total_profit2 = df[df["تم التسليم"] == True].groupby("جهة التمويل")["الربح"].sum()

comparison2 = pd.DataFrame({
    "عدد السيارات المباعة": sales_count2,
    "إجمالي الربح": total_profit2
})

comparison2
```

```python
filtered_df = df[
    (df["تم التسليم"] == True) &
    (df["جهة التمويل"] != "لايوجد")
]

sales_count2 = filtered_df.groupby("جهة التمويل")["تم البيع"].count()

total_profit2 = filtered_df.groupby("جهة التمويل")["الربح"].sum()

comparison2 = pd.DataFrame({
    "عدد السيارات المباعة": sales_count2,
    "إجمالي الربح": total_profit2
})
comparison2 = comparison2.drop("لايوجد", errors="ignore")
comparison2 = comparison2.drop("لايوجد, بنك الراجحي", errors="ignore")


comparison2
```

```python
comparison2.to_excel("مقارنه اداء كل جهه تمويل من حيث عدد المبيعات والربحية.xlsx", engine="openpyxl")
```

### حصر الأنواع أو الموديلات التي تحقق خسائر

```python
loss_cars = df.groupby("النوع")["الربح"].sum()
loss_cars = loss_cars[loss_cars < 0]

loss_cars = loss_cars.sort_values(ascending=True)

print("السيارات التي تحقق خسائر مرتبه من الاكثر الى الاقل  :")
loss_cars
 
```

```python
loss_cars.to_excel("حصر الأنواع أو الموديلات التي تحقق خسائر.xlsx", engine="openpyxl")
```

### تحليل المصدر الاكثر فعالية
